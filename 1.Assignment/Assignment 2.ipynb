## **Spliting the data into dependent and independent variables.**

```python
#splitting the data into dependent and independent feature
x=data.drop(columns='species',axis=1)
y=data.species
```

```python
print(x)
```

```python
print(y)
```

## **Perform Scaling**

```python
#importing the standard scaler library
from sklearn.preprocessing import StandardScaler
```

```python
Scaler=StandardScaler()
```

```python
#fitting the StandardScaler to the required data
x=Scaler.fit_transform(x)
print(x)
```

## **Spliting the data into training and testing**

```python
#importing the tain test split library from scikit learn
from sklearn.model_selection import train_test_split
```

```python
#splitting the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

## **checking which model is best for evaluation**

```python
#mportig some classification models and accuracy score libraries
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from  sklearn.metrics import accuracy_score
```

```python
models=[LogisticRegression(max_iter=10000),SVC(),KNeighborsClassifier(),RandomForestClassifier(),DecisionTreeClassifier()]
```

```python
for model in models:
  model.fit(x_train,y_train)
  acc=model.predict(x_train)
  acc_score=accuracy_score(acc,y_train)
  acc_score=acc_score*100
  acc_score=round(acc_score,2)
  print('accuracy_score on training data of',model,'=',acc_score)
```

```python
for model in models:
  model.fit(x_train,y_train)
  acc=model.predict(x_test)
  acc_score=accuracy_score(acc,y_test)
  acc_score=acc_score*100
  acc_score=round(acc_score,2)
  print('accuracy_score on testing data of',model,'=',acc_score)
```

**conclusion**:
Based on the accuracy scores you provided for both training and testing data, here are some conclusions and suggestions:
